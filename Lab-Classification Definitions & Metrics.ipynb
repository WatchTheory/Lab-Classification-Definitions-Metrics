# Conceptual

**Q:** Define regression.

> **Answer:**  Regression is a statistical method used in machine learning to model the relationship between a dependent variable and one or more independent variables. The goal is to predict a continuous output value.

**Q:** Define classification.

> **Answer:**  Classification is a machine learning task where the goal is to predict a discrete class label for a given input. It involves categorizing data into predefined classes.

**Q:** Define confusion matrix.

> **Answer:** A confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known. It allows for the visualization of the performance of an algorithm.

**Q:** Define `true positives`.

>**Answer:** refer to cases where the model correctly predicts the positive class.

**Q:** Define `true negatives`.

> **Answer:** True positives refer to cases where the model correctly predicts the positive class.

**Q:** Define `false positives`.

> **Answer:** refer to cases where the model incorrectly predicts the positive class when the actual class is negative.

**Q:** Define `false negatives`.

>**Answer:** are cases where the model incorrectly predicts the negative class when the actual class is positive.

# Scenarios

**Q: Fraud Detection System**
- Your team is developing a fraud detection system for online transactions. What classification metric should you primarily focus on to minimize the number of false positives, ensuring that legitimate transactions are not mistakenly flagged as fraudulent?

> **Answer:** to minimize the number of legitimate transactions mistakenly flagged as fraudulent (false positives), you should primarily focus on Precision. High precision means that when the system predicts fraud, it is very likely to be correct, thus reducing the inconvenience for legitimate users.

**Q: Medical Diagnosis**
- In a medical diagnosis model, which classification metric would be crucial to prioritize if the consequence of missing a positive case (e.g., a disease) is severe and should be minimized at all costs?
> **Answer:**
if the consequence of missing a positive case (a 'false negative') is severe and should be minimized at all costs, the crucial metric to prioritize is Sensitivity (also known as Recall or True Positive Rate). High sensitivity ensures that a high proportion of actual positive cases are correctly identified, reducing the risk of missing critical diagnoses.

**Q: Email Spam Filter**
- As part of enhancing an email spam filter, which classification metric should you emphasize to reduce the likelihood of legitimate emails being incorrectly marked as spam, thus avoiding disruptions to normal communication?
> **Answer:**
 to reduce the likelihood of legitimate emails being incorrectly marked as spam (which are false positives), you should emphasize Precision. A high precision ensures that when an email is flagged as spam, it is highly likely to actually be spam, minimizing disruption to normal communication for users.

**Q: Quality Control in Manufacturing**
- You are implementing a model for quality control in manufacturing, where identifying defective products is critical. Which classification metric would be most relevant for ensuring that the model effectively identifies all defective items with minimal false negatives?
> **Answer:**
 where identifying defective products is critical and minimizing false negatives is important, the most relevant classification metric is Sensitivity (also known as Recall or True Positive Rate). High sensitivity ensures that a high proportion of actual defective items are correctly identified, preventing faulty products from reaching consumers.

**Q: Credit Scoring System**
- Developing a credit scoring system for loan approval, which classification metric should be a key focus to ensure that the model minimizes the risk of approving a loan to someone who later defaults?
> **Answer:**
 to ensure that the model minimizes the risk of approving a loan to someone who later defaults, you should primarily focus on Precision. A high precision score means that when the system predicts a loan applicant will not default, it is highly likely to be correct, thus reducing financial losses for the lender.
